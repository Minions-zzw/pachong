```python
import requests
import time
from bs4 import BeautifulSoup
def download_page(url):
    headers = {'User-Agent': 'User-Agent:Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.87 Safari/537.36'}
    r = requests.get(url, headers=headers)  # 增加headers, 模拟浏览器
    print("OK")
    return r.text
```

```python
def save_txt(*args):
    for i in args:
        with open('qiubai.txt', 'a', encoding='utf-8') as f:
            f.write(i)
```

```python
def get_long_content(i): #获取长内容的文字
    href = i.find('a', class_='contentHerf').get('href')
    url = "https://www.qiushibaike.com{}".format(href)
    html = download_page(url)
    soup = BeautifulSoup(html, 'html.parser')
    string = soup.find('div',class_ = 'content').get_text()
    return string
```

```python
def get_content(html,page):
    output = """第{}页 作者：{} 性别：{} 年龄：{} 点赞：{} 评论：{}\n{}\n------------\n"""
    soup = BeautifulSoup(html, 'html.parser')
    con_list = soup.find_all('div', class_="article")
    for i in con_list:
        author_name = i.find('h2').string#获取作者名字
        author_age = i.find('div', class_ = 'articleGender').string #获取作者年龄
        author_gender = '男' if(i.find('div', class_ = 'articleGender').get('class')[1] == 'manIcon') else '女' #获取作者年龄
        if(i.find('span', class_ = 'contentForAll')):#看是否有查看全文
            time.sleep(3)
            content = get_long_content(i)#从新页面获取长内容
        else:
            content = i.find('div', class_ = 'content').find('span').get_text()
#         content = i.find('div', class_ = 'content').find('span').get_text()
        vote = i.find('span', class_ = 'stats-vote').find('i').string
        comment = i.find('span', class_ = 'stats-comments').find('i',class_ = 'number').string
        save_txt(output.format(page, author_name, author_gender, author_age, vote, comment, content))
```

```python
#main函数
for i in range(1,4):
    url = "https://www.qiushibaike.com/text/page/{}".format(i)
    print(i)
    html = download_page(url)
    get_content(html, i)
```
